#### Loading the datasets into the notebook

```python
Ratingsdf =  pd.read_csv ("Netflix_Dataset_Rating.csv")
Ratingsdf.head()
```

```python
Moviesdf = pd.read_csv ("Netflix_Dataset_Movie.csv")
Moviesdf.head()
```

```python
MoviesInfo_df = pd.read_csv ("imdb_processed.csv")
MoviesInfo_df.head()
```

```python
# Create one merged DataFrame containing all the Netflix data.
MoviesDF = Moviesdf.merge(Ratingsdf, on='Movie_ID').merge(MoviesInfo_df, on=['Name','Year'])
MoviesDF.head()
```

```python
new_MoviesDF = MoviesDF.drop(columns=['kind','vote','country','cast','composer','runtime'])
new_MoviesDF.head()
```

```python
##The information on the number of rows and columns present in our dataframe
new_MoviesDF.shape
```

```python
## The information on the dataset
new_MoviesDF.info()
```

```python
## Missing information within the dataframe
new_MoviesDF.isnull().sum()
```

```python
#The output is the number of unique users in the ratings dataframe
unique_users = Ratingsdf['User_ID'].nunique()
print(unique_users)
```

```python
#The output is the number of unique users in the new movies dataframe
countunique_users = new_MoviesDF['User_ID'].nunique()
print(countunique_users)
```

```python
#The output is the number of unique movies in the movies dataframe
unique_movies = Moviesdf['Movie_ID'].nunique()
print(unique_movies)
```

```python
#The output is the number of unique movies in the new movies dataframe
countunique_movies = new_MoviesDF['Movie_ID'].nunique()
print(unique_movies)
```

```python
##Getting the ratings per user for at least 10 unique users 
users_ratings = (
    new_MoviesDF
    .groupby('User_ID', as_index=False)
    .agg({'Rating': ['count', 'mean']})
)
users_ratings.head(10)
```

```python
## Getting the ratings per movie by the ratings 
movies_ratings = (
    new_MoviesDF
    .groupby('Movie_ID', as_index=False)
    .agg({'Rating': ['count', 'mean']})
)
movies_ratings.head(10)
```
